# TEOREMAS

---

**Teorema de la resta de dos variables aleatorias**

---

Si \( X \) y \( Y \) son variables aleatorias, entonces:

* La esperanza de la resta es:

  $$
  \mathbb{E}[X - Y] = \mathbb{E}[X] - \mathbb{E}[Y]
  $$

* Si \( X \) y \( Y \) son independientes, la varianza de la resta es:

  $$
  \mathrm{Var}(X - Y) = \mathrm{Var}(X) + \mathrm{Var}(Y)
  $$
  
*  Si \( $X \sim \mathcal{N}$($\mu_X, \sigma_X^2$) \) y \($ Y \sim \mathcal{N}$($\mu_Y, \sigma_Y^2$) \), entonces:

  $$
  X - Y \sim \mathcal{N}(\mu_X - \mu_Y,\ \sigma_X^2 + \sigma_Y^2)
  $$

  ---

---

**Teorema del Límite Central (TLC):**

---

Sea \( $X_1, X_2, \dots, X_n $\) una sucesión de variables aleatorias independientes e idénticamente distribuidas (i.i.d.) con media \($ \mu $\) y varianza \($ \sigma^2 $\), entonces la suma estandarizada:

$$
Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}}
$$

donde \( $\bar{X}n = \frac{1}{n} \sum{i=1}^{n} X_i $\), converge en distribución a una normal estándar:

$$
Z_n \xrightarrow{d} \mathcal{N}(0, 1) \quad \text{cuando } n \to \infty
$$

Esto implica que, para valores suficientemente grandes de \( n \), la distribución de \( $\bar{X}_n$ \) se aproxima a una distribución normal:

$$
\bar{X}_n \sim \mathcal{N}\left(\mu,\ \frac{\sigma^2}{n} \right)
$$

***Nota:*** Este resultado es válido incluso si la distribución original de las \($ X_i $\) no es normal, siempre que tengan media y varianza finitas.

**Teorema de la probabilidad del producto de dos variables aleatorias**

---

Sea \($ X $\) y \($ Y $\) dos variables aleatorias.

*  La probabilidad conjunta de que \( X \) tome el valor \( x \) y \( Y \) el valor \( y \) es:

  $$
  \mathbb{P}(X = x y  Y = y) = \mathbb{P}(X = x, Y = y)
  $$

* Si \( X \) y \( Y \) son independientes, entonces:

  $$
  \mathbb{P}(X = x, Y = y) = \mathbb{P}(X = x) \cdot \mathbb{P}(Y = y)
  $$

* En el caso continuo, si \( $f_X(x)$ \) y \( $f_Y(y)$ \) son las funciones de densidad marginales, entonces si \($ X $\) y \($ Y $\) son independientes:

  $$
  f_{X,Y}(x, y) = f_X(x) \cdot f_Y(y)
  $$

  y la probabilidad de que \($ X \in A $\) y \( $Y \in B $\) es:

  $$
  \mathbb{P}(X \in A,\ Y \in B) = \int_A \int_B f_X(x) \cdot f_Y(y) \, dy \, dx
  $$

## Uniforme

$\mathcal{U}(a,b) \qquad a<b$

$f(x)=$  
\begin{cases}
    \frac{1}{b-a} & \text{para } a \leq x \leq b \\
    0 & \text{en otro caso}
\end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/500px-Uniform_Distribution_PDF_SVG.svg.png" width = "250"/>
</div>

$F(x)=$
\begin{cases}
    0 & \text{para } x < a \\
    \frac{x-a}{b-a} & \text{para } a \leq x \leq b \\
    1 & \text{para } x > b
\end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/63/Uniform_cdf.svg/500px-Uniform_cdf.svg.png" width = "250"/>
</div>


$$E[X]=\frac{a + b}{2}$$

$$V[X]=\frac{(b - a)^2}{12}$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUCODÓDIGO**
1. Generar $U\sim U(0,1)$
2. Regresar $X=Ub+a(1-U)$

```python
import random
import matplotlib.pyplot as plt
def unif(a,b,n ):
  x=[] #lista para almacenar las desviciones generadas
  for i in range(n): #ciclo para generar los números aleatorios
    R=random.random()
    x.append((R*b)+a*(1-R)) #almacenar las desviaciones en la lista
  return x

exper = unif(1,6,10000)

plt.hist(exper, bins=30,color = "pink")
plt.title("Uniforme f(x)")
plt.show()
print('MEDIA:', np.mean(exper))
print('VARIANZA:', np.var(exper))
print('DESVIACIÓN ESTÁNDAR:', np.std(exper))
```

## Triangular

$\mathcal{Tri(a,b, c)}$

$f(x) =$ \begin{cases}
    \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
    \frac{2(b-x)}{(b-a)(b-c)} & \text{para } c \leq x \leq b \\
    0 & \text{en otro caso}
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/45/Triangular_distribution_PMF.png/330px-Triangular_distribution_PMF.png" width = "250"/>
</div>

  $F(x)=$ \begin{cases}
    0 & \text{para } x < a \\
    \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
    1 - \frac{(b-x)^2}{(b-a)(b-c)} & \text{para } c \leq x \leq b \\
    1 & \text{para } x > b
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/e5/Triangular_distribution_CMF.png/330px-Triangular_distribution_CMF.png" width = "250"/>
</div>

  $$E[X]=\frac{a+b+c}{3}$$

  $$V[X]=\frac{a^2+b^2+c^2-ab-ac-bc}{18}$$

  $$\sigma[X] = \sqrt{V[X]}$$

  
  **PSEUDOCÓDIGO**
  1. Generar $U\sim U(0,1)$
  2. Retornar:
  $$x=\begin{cases} \sqrt{U(c-a)(b-a)}+a & si & U<\frac{b-a}{c-a}\\
\\
c-\sqrt{(1-U)(c-a)(c-b)} & si & U\geq \frac{b-a}{c-a}
\end{cases}$$

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math

def tri(a,b,c,n ):
    x=[]#lista vacía para almacenar las x calculadas
    for i in range(n):#iniciamos ciclo para repetir el proceso n veces
       R = random.random()#generamos el número aleatorio
       xi = np.sqrt(R*(c-a)*(b-a))+a #operación del primer valor de x
       if a <= xi <= b: # definimos el condicional
          x.append(xi) # si cumple las condiciones se almacena en la lista
       else:# sino las cumple pasamos a clcular el otro valor de x
         xi = c-np.sqrt((1-R )*(c-a )*(c-b ))
         x.append(xi)
    return x #regresa la lista de los n valores de x calculados

#definimos los parámetros iniciales
a=1
b=3
c=5
#las veces que se repetirá el experimento
n=10000
#asignamos una variable al experimento
h =tri(a,b,c,n)
#graficamos nuestra función de distribución Triangular
plt.hist(h, bins=40, color = "pink")
plt.title("Triangular f(x)")
plt.show()
print('MEDIA:', np.mean(h))
print('VARIANZA:', np.var(h))
print('DESVIACIÓN ESTÁNDAR:', np.std(h))
```

## Exponencial


$\mathcal{Exp(\lambda)} \qquad \lambda > 0 \qquad x\in [0,\infty)$

$ f(x) =$ \begin{cases}
    \lambda e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/ec/Exponential_pdf.svg/800px-Exponential_pdf.svg.png" width = "250"/>
</div>

$F(x)=$ \begin{cases}
    1 - e^{-\lambda x} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/b/ba/Exponential_cdf.svg/800px-Exponential_cdf.svg.png" width = "250"/>
</div>

  $$E[X]=\frac{1}{\lambda}$$
  
  $$V[X]=\frac{1}{\lambda^2}$$

  $$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCODIGO  INVERSA**
1. Generar U $\sim$ U(0, 1)
2. Salir  X = -$\frac{ln(R)}{\lambda}$

```python
import random
import matplotlib.pyplot as plt
import numpy as np

def exponencial(la,n ):
  x=[]#lista para almacenar las desviciones generadas
  for i in range(n):#ciclo para generar los números aleatorios
    R=random.random()
    x.append(-np.log(R)/la)#almacena cada desviación en la lista
  return x

A=exponencial(1,10000)
print('MEDIA:', np.mean(A))
print('VARIANZA:', np.var(A))
print('DESVIACIÓN ESTÁNDAR:', np.std(A))

plt.hist(A, bins=50,color = "pink")
plt.title("Exponencial f(x)")
plt.show()
```

## m-Erlang

$\mathcal{Erlang(\lambda, n)} \qquad \lambda >0 \qquad n \in \mathbb{Z}^+ \qquad x \in [0,\infty)$
  
$f(x) =$ \begin{cases}
    \frac{\lambda (\lambda x)^{n-1} e^{-\lambda x}}{(n-1)!} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/f/fc/Gamma_distribution_pdf.png/800px-Gamma_distribution_pdf.png" width = "250"/>
</div>

$F(x) = 1 - \sum_{k=0}^{n-1} \frac{(\lambda x)^k e^{-\lambda x}}{k!}$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a9/Gamma_distribution_cdf.png/800px-Gamma_distribution_cdf.png" width = "250"/>
</div>

  $$E[X] = \frac{n}{\lambda}$$


  $$V[X] = \frac{n}{\lambda^2}$$

  $$\sigma[X] = \sqrt{V[X]}$$

  
  **PSEUDOCÓDIGO**
  1. Generar $U_1,U_2,...,U_m$ como IID (Independiente e Independientemente distribuido) U(0,1)

2. Return $X=-\frac{\mu}{m}ln(\prod_{i=1}^{m}U_i)=\frac{ln(\prod_{i=1}^{m}U_i)}{\lambda}$

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

def erlang(m, lanbda,n):
  x = []#lista para almacenar las desviciones generadas
  mu = m / lanbda #Media
  for i in range(n):#ciclo para generar los números aleatorios
    U = [random() for _ in range(m)]#ciclo  para obtener m números aleatorios
    producto = np.prod(U)#productario de los U_i
    x.append(-(mu / m) * np.log(producto)) #almacena cada desviación en la lista
  return x

exp = erlang(3, 2,10000)
print("Media aprox:",np.mean(exp))
print("Varianza aprox: ",np.var(exp))
print("Desviación estandar aprox: ",np.std(exp))
plt.hist(exp, bins=100,color = "pink")
plt.title("m-Erlang f(x)")
plt.show()
```

### Gamma 1

$\mathcal{Gamma(α, β)} \qquad \alpha, \beta >0 \qquad x \in [0,\infty )$

Tomando $m=\alpha \qquad λ=\frac{1}{β}$

$\frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} ⇒ \frac{\lambda^m x^{m-1} e^{-\lambda x}}{\Gamma(m)}$

$f(x) =$ \begin{cases}
    \frac{\beta^{-\alpha} x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

con $\Gamma(k+1)=k!$

$F(x)=$ \begin{cases}
    1- e^{-\frac{x}{\beta}}\sum_{i=0}^{\alpha -1}\frac{(\frac{x}{\beta})^j}{j!} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

  $$E[X] = \alpha \beta$$


  $$V[X] = \alpha \beta^2$$

  $$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO 1**

Para $0<\alpha< 1$:

con $b=\frac{e+\alpha}{e}$

1. Generar $U_1$~$U(0,1)$ y sea $P=bU_1$
  
  SI P>1, ir al paso 3
  
  otro caso: ir al paso 2

2. Sea $Y=P^{\frac{1}{\alpha}}$ y generar $U_2$~$U(0,1)$
   
   SI $U_2\leq e^{-Y}$

      return $X=Y$
   otro caso: ir al paso 1.

3. Sea $Y=-ln(\frac{b-P}{\alpha})$ y generar $U_2$~$U(0,1)$

    SI $U_2\leq Y^{\alpha-1}$

      return $X=Y$

    otro caso: ir al paso 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
#0<alpha<1 y beta=1
def gamma1(al,n):
    x=[]#lista qque almacena las desviaciones generadas
    #Parámetros
    e = np.e
    b = (e + al) / e
    for _ in range(n):#ciclo para generar los números aleatorios
        U1=random.random()
        P=b*U1#calcular el valor de P
        if P>1:#primer condcional sobre Y
            Y=-np.log((b-P)/al)
            U2=random.random()
            if U2<=Y**(al-1):#segundo condicional sobre Y
                x.append(Y)
        else:#en otro caso
            Y=P**(1/al)
            U2=random.random()
            if U2<=e**(-Y):
                x.append(Y)
    return x
G=gamma1(0.5,10000)
print("Media aprox:",np.mean(G))
print("Varianza aprox: ",np.var(G))
print("Desviación estandar aprox: ",np.std(G))
plt.hist(G, bins=100, color = "pink")
plt.title("Gamma 1 f(x)")
plt.show()
```

### Gamma 2

Para $\alpha>1$

$\beta=\frac{1}{\sqrt{2\alpha -1}}$

$b=\alpha-ln(4)$

$q=\alpha+\frac{1}{a}$

$\theta =4.5$

$d=1+ln(θ)$

  $$E[X] = \alpha \beta$$


  $$V[X] = \alpha \beta^2$$

  $$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO 2**

1. Generar $U_1,U_2$ como IID $U(0,1)$

2. Sea $V=a\cdot ln(\frac{U_1}{1-U_1}) \qquad Y=\alpha e^V \qquad Z=U_1^2U_2 \qquad W=b+qV-Y$

3. SI $W+d-θZ\geq 0$

   return $X=Y$

  otro caso: ir al paso 4

4. SI $W\geq ln(Z)$
  return $X=Y$

  otro caso: ir al paso 1

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def gamma2(al,n):
  if (0 < al < 1):#condicional sobre el valor de alpha
    raise ValueError("Válido para alpha > 1.")
#parámetros
  be = 1 / np.sqrt(2 * al - 1)
  b = al - np.log(4)
  q = al + 1 / be
  theta = 4.5
  d = 1 + np.log(theta)

  x = []
  for _ in range(n):#ciclo para generar los números aleatorios
    U1 = random.random()
    U2 = random.random()
    #cálculo de las variables
    V = be * np.log(U1 / (1 - U1))
    Y = al * np.exp(V)
    Z = U1 ** 2 * U2
    W = b + q * V - Y
    if W + d - theta * Z >= 0:#primer condicinal
      x.append(Y)
    elif W >= np.log(Z):#en otro caso
      x.append(Y)
  return x

G2 = gamma2(2,10000)
plt.hist(G2, bins=100, color = "pink")
plt.title("Gamma 2 f(x)")
plt.show()
print("Media aprox:",np.mean(G2))
print("Varianza aprox: ",np.var(G2))
print("Desviación estandar aprox: ",np.std(G2))
```

## Weibull

$\mathcal{We(\alpha, \beta)} \qquad \alpha, \beta >0 \qquad x \in (0, \infty)$

$f(x)=$\begin{cases}
    \alpha \beta^{-\alpha}x{\alpha - 1}e^{-(x/\beta)^\alpha} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

$= \lambda\alpha(\lambda x)^{\alpha-1}e^{(\lambda x)\alpha}$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/5/58/Weibull_PDF.svg/800px-Weibull_PDF.svg.png" width = "250"/>
</div>

$F(x)=$\begin{cases}
    1-e^{-(\frac{x}{\beta})^\alpha} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
$= 1-e^{-(\lambda x)^\alpha}$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/7/7e/Weibull_CDF.svg/800px-Weibull_CDF.svg.png" width = "250"/>
</div>

$$E[X]=\frac{\beta}{\alpha}\Gamma(\frac{1}{\alpha})=\frac{1}{\lambda}\Gamma(1+\frac{1}{\alpha})$$

$$V[X]=\frac{\beta^2}{\alpha}[2\Gamma(\frac{2}{\alpha})-\frac{1}{\alpha}[\Gamma(\frac{1}{\alpha})]^2]= \frac{1}{\lambda^2}[\Gamma(1+\frac{2}{\alpha})-\Gamma^2(1+\frac{1}{\alpha})]$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**
1. Generar $U$~$U(0,1)$
2. Return $X=β(-ln(U))^{\frac{1}{\alpha}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def weibull(be,al, n):
  x = []#lista para almacenar las desviaciones generadas
  for _ in range(n):#ciclo para generar los números aleatorios
    U = random.random()
    x.append(be*(-np.log(U))**(1/al))#calculo de Fˆ-1
  return x

w = weibull(1,3,10000)
plt.hist(w, bins=100,color = "pink")
plt.title("Weibull f(x)")
plt.show()
print("Media aprox:",np.mean(w))
print("Varianza aprox: ",np.var(w))
print("Desviación estandar aprox: ",np.std(w))
```

## Normal

$\mathcal{N(\mu, \sigma^2)} \qquad \mu \in (-∞, ∞) \qquad \sigma>0 \qquad x \in(-∞, ∞)$

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/1b/Normal_distribution_pdf.png/800px-Normal_distribution_pdf.png" width = "250"/>
</div>

$F(x)$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/19/Normal_distribution_cdf.png/800px-Normal_distribution_cdf.png" width = "250"/>
</div>

$$E[X]=\mu$$

$$V[X]=\sigma^2$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO BOX-MULLER**

1. Generar $U_1, U_2  \sim U(0,1)$

2. Hacer: $\quad R=\sqrt{-2lnU_1}\qquad , \qquad \theta=2\pi  U_2$

3. Hacer: $\quad X=Rcos(\theta) \qquad ,\quad Y=Rsen(\theta)$

4. Salida: $X,Y$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def BM(n):
    lista=[]#lista para almacenar las desviaciones
    for _ in range(n):#ciclo para generar los números aleatorios
        u1=rd.random()
        u2=rd.random()
        #Parámetros
        r = np.sqrt(-2*np.log(u1))
        theta=2*np.pi*u2
        x=r*np.cos(theta)
        y=r*np.sin(theta)
        lista.append(x)#almacena la desviación x
        lista.append(y)#almacena la desviación y
    return lista
BM = BM(10000)
print("Media aprox:",np.mean(BM))
print("Varianza aprox: ",np.var(BM))
print("Desviación estandar aprox: ",np.std(BM))
plt.hist(BM, bins=100,color = "pink")
plt.title("Normal f(x)")
plt.show()
```

## Log-normal

$\mathcal{LN}(\mu,\sigma^2) \qquad \sigma>0 \qquad e^\mu >0 \qquad x \in (-∞, ∞)$

$f(x) =$\begin{cases}
    \frac{1}{\sigma x \sqrt{2\pi}} e^{-\frac{(ln \quad x-\mu)^2}{2\sigma^2}} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/8/89/Log-normal-pdfs.png/330px-Log-normal-pdfs.png" width = "250"/>
</div>

$F(x)$

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Log-normal-cdfs.png/330px-Log-normal-cdfs.png" width = "250"/>
</div>

$$E[X]=e^{\mu + \sigma^2 /2}$$

$$V[X]=e^{\mu + \sigma^2 /2}(e^{\sigma^2}-1)$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**

con $E[X]=\mu'= e^{\mu + \frac{\sigma^2}{2}} \qquad V[X]=\sigma'^2=(e^{\sigma^2} - 1) e^{2\mu + \sigma^2}$

Entonces

$E[Y]=\mu=ln( \frac{\mu´^2}{\sqrt{\mu´^2+\sigma´^2}})$

$V[Y]\sigma^2=ln(1+\frac{\sigma´^2}{\mu´^2})$

1. Generar $Y \sim N(\mu, \sigma^2)$
2. return $X=e^Y$



```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def log_normal(mup,sigp,n):
  #Parámetros
  mu=np.log(mup**2/np.sqrt(mup**2+sigp**2))
  sigma=np.sqrt(np.log(1+sigp**2/mup**2))
  X=[]
  for _ in range(n):#ciclo para generar los números aleatorios
    Y=np.random.normal(mu,sigma)
    X.append(np.exp(Y))#almacen de la desviación generada
  return X

L=log_normal(1,1,1000)
plt.hist(L, bins=100,color = "pink")
plt.title("Log-normal f(x)")
plt.show()
print("Media aprox:",np.mean(L))
print("Varianza aprox: ",np.var(L))
print("Desviación estandar aprox: ",np.std(L))
```

  $$E[X]=\frac{\alpha}{\alpha + \beta}$$

$$V[X]=\frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**

1. Generar $Y_1\sim$gamma($\alpha$,1) y $Y_2\sim$gamma($\beta$,1) independiente de $Y_1$
2. Regresar $X=\frac{Y_1}{Y_1+Y_2}$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def beta(al,be,n):
  X=[] #lista para almacenar las desviaciones
  for _ in range(n):#ciclo para generar los números aleatorios
    Y1=np.random.gamma(al,1)
    Y2=np.random.gamma(be,1)
    X.append(Y1/(Y1+Y2))#almacen de las X generadas
  return X

k=beta(0.5,0.5,1000)
plt.hist(k, bins=50,color = "pink")
plt.title("Beta f(x)")
plt.show()
print("Media aprox:",np.mean(k))
print("Varianza aprox: ",np.var(k))
print("Desviación estandar aprox: ",np.std(k))
```

$\mathcal{PT6(\alpha_1, \alpha_2, \beta)} \qquad \alpha_1,\alpha_2,\beta>0$

Con $F_B$la función de distribución beta $\beta(\alpha, \frac{1}{\beta})$

1. Generar $Y_1\sim gamma(\alpha_1,\beta) \qquad Y_2\sim gamma(\alpha_2,1)$
   
   donde $Y_1$ y $Y_2$ son independientes
2. Return $X=\frac{Y_1}{Y_2}$

```python
import numpy as np
import matplotlib.pyplot as plt

def pearson_vi(alpha1,alpha2,beta,n):
  x=[]#lista para almacenar
  for _ in range(n):#ciclo para generar los números aleatorios
    Y1=np.random.gamma(alpha1,beta)
    Y2=np.random.gamma(alpha2,1)
    x.append(Y1/Y2)#generar y guardar los X
  return x

P6 = pearson_vi(2,4,1,10000)
plt.hist(P6, bins=100,range=(0,5),color = "pink")
plt.title("PT6 f(x)")
plt.show()
print("Media aprox:",np.mean(P6))
print("Varianza aprox: ",np.var(P6))
print("Desviación estandar. ", np.std(P6))
```

## Johnson


### Bounded

$\mathcal{JSB(\alpha_1,\alpha_2,a,b)} \qquad a \in(-∞,+∞) \quad b>a  \quad \alpha_1\in(-∞,+∞) \quad \alpha_2>0$

  $
  f(x) = \begin{cases}
    \frac{\alpha_2(b-a)}{(x-a)(b-x)\sqrt{2\pi}}e^{-\frac{1}{2}[\alpha_1+\alpha_2ln(\frac{x-a}{b-x})]} & \text{si } a<x<b \\
    0 & \text{otro caso}
  \end{cases}
  $
1000026243.jpg1000026244.jpg

$
  F(x) = \begin{cases}
    Φ [\alpha_1+\alpha_2ln(\frac{x-a}{b-x})] & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
$

donde $𝚽(x)$ es la distribución Normal con $\mu=0 \quad \sigma^2=1$

No hay media y varianza establecida

**PSEUDOCÓDIGO**

$X\sim JSB(\alpha_1,\alpha_2,a,b) \iff Z=\alpha_2+\alpha_2 ln[\frac{X-a}{b-X}] \sim N(0,1)$
1. generar $Z\sim N(0,1)$

2. Hacer $Y=e^{\frac{Z-\alpha_1}{α_2}}$

3. return $X=\frac{(a+bY)}{(Y+1)}$

```python
import numpy as np
import matplotlib.pyplot as plt

def johnson(alpha1,alpha2,a,b,n):
  x=[]#lista para almacenar los datos
  for _ in range(n):#ciclo para generar los números aleatorios
    Z=np.random.normal(0,1)
    Y=np.exp((Z-alpha1)/alpha2)
    x.append((a+b*Y)/(Y+1))#generar y guardar los  X
  return x

JB = johnson(0,1,0,1,10000)

plt.hist(JB, bins=100,range=(0,1), color = "pink")
plt.title("JSB f(x)")
plt.show()
print("Media aprox:",np.mean(JB))
print("Varianza aprox: ",np.var(JB))
```

## Uniforme

$\mathcal{U(n)}$

$$f(x)=\frac{1}{n} \qquad x=1,2,\cdots,n$$

</div>

$$E[x]=\frac{n+1}{2}$$

$$V[x]=\frac{n^2-1}{12}$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCODIGO  INVERSA**
1. Generar U $\sim$ U(0, 1)
2. Salir  X = a + (b - a)U

$\mathcal{DU(a,b)}$

La variable aleatoria \( X \) toma valores en el conjunto ${a, a+1, \dots, b}$, donde  $a \leq b$ y $ a, b \in \mathbb{Z} $.

$f(x = k)$ = \begin{cases}
\dfrac{1}{b - a + 1} & \text{si } k \in \{a, a+1, \dots, b\}, \\[0.8em]
0 & \text{en otro caso.}
\end{cases}


$F(x) =$
\begin{cases}
0 & \text{si } x < a, \\[0.5em]
\dfrac{\lfloor x \rfloor - a + 1}{b - a + 1} & \text{si } a \leq x \leq b, \\[0.5em]
1 & \text{si } x > b,
\end{cases}

donde $ \lfloor x \rfloor $ denota la **función piso** (el mayor entero menor o igual a x).

$$E[x]=\frac{a+b}{2}$$

$$V[x]=\frac{(b-a+1)^2-1}{12}$$

$$\sigma[X] = \sqrt{V[X]}$$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def uniforme(n, a, b):
    unif = []#lista para almacenar los X
    for _ in range(n):#ciclo para generar los números aleatorios
        R = rd.uniform(a, b)
        x = a + (b-a)*R
        unif.append(x)#guarda los x calculados
    return unif

A=uniforme(10000, 0, 1)
print('MEDIA:', np.mean(A))
print('VARIANZA:', np.var(A))
print('DESVIACIÓN ESTÁNDAR:', np.std(A))
plt.hist(A, bins=10, color = "pink")
plt.title("Uniforme f(x)")
plt.show()
```

## Bernoulli

$\mathcal{B(p)} \qquad p \in (0,1)$

<div align="center">
<img src = "https://www5.uva.es/estadmed/probvar/d_univar/bernoulli.gif" width = "450"/>
</div>

$f(x)$ =
\begin{cases}
p       & \text{si } x = 1, \\[0.5em]
1 - p   & \text{si } x = 0, \\[0.5em]
0       & \text{en otro caso.}
\end{cases}

$F(x)$ =
\begin{cases}
0               & \text{si } x < 0, \\[0.5em]
1 - p           & \text{si } 0 \leq x < 1, \\[0.5em]
1               & \text{si } x \geq 1.
\end{cases}

$$E[x]=p$$

$$V[x]=p(1-p)$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**
1. Generar $U\sim U(0,1)$
2. Si $U\leq p$

  return X=1

  Otro caso

  return X=0

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def bernoulli(n, p):
    bern = []#lista para almacenar los x
    for _ in range(n):#ciclo para generar los números aleatorios
        R = rd.uniform(0, 1)
        if R <= p:#primer condicion sobre R
            x = 1
        else:#en otro caso
            x = 0
        bern.append(x)#almacena el resultado de las condiciones
    return bern

BER=bernoulli(10, 0.5)
print('MEDIA:', np.mean(BER))
print('VARIANZA:', np.var(BER))
print('DESVIACIÓN ESTÁNDAR:', np.std(BER))
plt.hist(BER, bins=50,color = "pink")
plt.title("Bernoulli f(x)")
plt.show()
```

## Geométrica

$\mathcal{geom(p)} \qquad p \in (0,1)$

<div align="center">
<img src = "https://www5.uva.es/estadmed/probvar/d_univar/geomet.gif" width = "450"/>
</div>

$f(x)=$
\begin{cases}
p(1-p)^x       & \text{si } x = 0, 1, \cdots  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}

$F(x)$ =
\begin{cases}
0 & \text{otro caso}, \\[0.5em]
1 - (1 - p)^{\lfloor x \rfloor +1} & \text{si } x \geq 0.
\end{cases}

$$E[x]=\frac{1-p}{p}$$

$$V[x]=\frac{1-p}{p^2}$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**
1. Generar $U\sim U(0,1)$
2. Return $X=[ln(U)/ln(1-p)]$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def geometrica(n, p):
    geom = []#lista vacía para almacenar los valores
    for _ in range(n):#ciclo para generar los números aleatorios
        R = rd.uniform(0, 1)
        geom.append(np.log(R)/np.log(1-p))#calcula y guarda los X
    return geom

GEOM=geometrica(1000, 0.5)
print('MEDIA:', np.mean(GEOM))
print('VARIANZA:', np.var(GEOM))
print('DESVIACIÓN ESTÁNDAR:', np.std(GEOM))
plt.hist(GEOM, bins=50, color = "pink")
plt.title("Geometrica f(x)")
plt.show()
```

## Binomial negativa

$\mathcal{negbin(s,p)} \qquad s>0 \qquad p \in (0,1)$

<div align="center">
<img src = "https://www5.uva.es/estadmed/probvar/d_univar/bin_neg.gif" width = "450"/>
</div>

$f(x)=$
\begin{cases}
\binom{s + x - 1}{x} p^s (1-p)^x       & \text{si } x = 0,1,2,\cdot  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}



$F(x)=$
\begin{cases}
\sum_{i=0}^{⌊x ⌋}\binom{s + i - 1}{i} p^s (1-p)^i       & \text{si } x ≥0  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}


$$E[x]=\frac{s(1-p)}{p}$$

$$V[x]=\frac{s(1-p)}{p^2}$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**
1. Generar $Y_1,Y_2,..,Y_s$ como IID geom(p) random variables.
2. Return $X=Y_1+Y_2+...+Y_s$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def binomial_negativa(n, s, p):
    binomial = [] #lista para almacenar los X
    for _ in range(n):#ciclo para generar los números aleatorios
        R=np.random.geometric(p, s)
        binomial.append(np.sum(R))#clacula y guarda los X
    return binomial

BN=binomial_negativa(10000, 10, 0.5)
plt.hist(BN, bins=50, color = "pink")
plt.title("Binomial negativa f(x)")
plt.show()
print('MEDIA:', np.mean(BN))
print('VARIANZA:', np.var(BN))
print('DESVIACIÓN ESTÁNDAR:', np.std(BN))
```

## Poisson

$\mathcal{p(\lambda)} \qquad \lambda >0 \qquad x \in {0,1,2,\cdots}$

$f(x)=$
\begin{cases}
\frac{e^{-\lambda} \lambda^x}{x!}       & \text{si } x = 0,1,2,\cdot  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/16/Poisson_pmf.svg/330px-Poisson_pmf.svg.png" width = "250"/>
</div>

$F(x)=$
\begin{cases}
0 & x < 0 \\[0.5em]
e^{-\lambda}\sum_{i=0}^{⌊x ⌋}\frac{\lambda ^i}{i!} & \text{si } x \geq 0.
\end{cases}

<div align="center">
<img src = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a2/PoissonCDF.png/330px-PoissonCDF.png" width = "250"/>
</div>

$$E[x]=\lambda$$

$$V[x]=\lambda$$

$$\sigma[X] = \sqrt{V[X]}$$

**PSEUDOCÓDIGO**
1. Sea $a=e^{-\lambda}, \quad b=1, \quad i=0$
2. Generar $U_{i+1}\sim U(0,1)$
3. Reemplazar  $bU_{i+1}$
4. Si $b<a$
   return $X=i$
   
   Otro caso: ir a paso 5
5. Reemplazar $i$ por $i+1$ y regresar al paso 2

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def poisson(lam, n_muestras):
    P = []##lista para almacenar los datos
    for _ in range(n_muestras):#ciclo para generar los números aleatorios
    #Parámetros
        a = math.exp(-lam)
        b = 1.0
        i = 0#contador de evento
        while True:
            u = np.random.uniform(0, 1)
            b *= u
            if b < a:# Cuando b cae por debajo de e^(-λ), se detiene
                break
            i += 1 # Cada vez que b >= a, se incrementa el contador i
        P.append(i)

    return P

p = poisson(4, 100)
print('MEDIA:', np.mean(p))
print('VARIANZA:', np.var(p))
print('DESVIACIÓN ESTÁNDAR:', np.std(p))
# Graficar histograma
plt.hist(p, bins=100, color = "pink")
plt.title("Poisson f(x)")
plt.show()
```
